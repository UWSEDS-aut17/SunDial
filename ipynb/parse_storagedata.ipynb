```python
import pandas as pd
import numpy as np
import matplotlib as plt
import xlrd
```

```python
stor_init_df = pd.read_csv('data/storage_init.csv')
stor_df = stor_init_df
df50 = stor_df.loc[stor_df['TEMP'] == 50]
df25 = stor_df.loc[stor_df['TEMP'] == 25]
dfn5 = stor_df.loc[stor_df['TEMP'] == -5]
dfn40 = stor_df.loc[stor_df['TEMP'] == -40]
```

```python
fiftyC_names = ['12_02_2014_IC_PLN_50C_1.xls',
            '12_17_2014_PLN_IC_50C_2.xls',
             '1_08_2015_PLN_IC_50C_3.xls',
             '1_30_2015_PLN_IC_50C_4.xls',
             '2_23_2015_PLN_IC_50C_5.xls',
             '3_16_15_PLN_IC_50C_6.xls',  
             '04_06_2015_PLN_IC_50C_6.xlsx',
             '04_29_2015_50C_3W.xls',
             '05_28_2015_50C_3W_Final.xls',
             '06_29_2015_50C_3W_IC.xls']

a = (0,1,2,3,8,9,10,11)
```

```python
for jj in range(len(fiftyC_names)):
    workbook = xlrd.open_workbook('data/Capacity_50C/3W/' + fiftyC_names[jj])
    df50[fiftyC_names[jj]] = float(0)
    if jj < 6:
        for ii in range(1,13):        
            worksheet = workbook.sheet_by_index(ii)    
            df50[fiftyC_names[jj]][ii-1] = worksheet.cell(-1,9).value
    else:
        for ii in range(len(a)):
            worksheet = workbook.sheet_by_index(ii+1)    
            df50[fiftyC_names[jj]][a[ii]] = worksheet.cell(-1,9).value
```

```python
df = stor_df
df['Elap_time'] = 0
df['Cap'] = 1



for jj in range(len(fiftyC_names)):
    idx = len(df)
    for ii in range(12):
        df.loc[idx+ii,:] = df.loc[ii,:]
        df.loc[idx+ii,'Elap_time'] = 21*(jj+1)
        df.loc[idx+ii,'Cap'] = df50.loc[ii,fiftyC_names[jj]]/df.loc[ii,'Discharge Capacity']
        
for jj in range(len(tfC_names)):
    idx = len(df)
    for ii in range(12):
        df.loc[idx+ii,:] = df.loc[ii+12,:]
        df.loc[idx+ii,'Elap_time'] = 21*(jj+1)
        df.loc[idx+ii,'Cap'] = df25.loc[ii+12,tfC_names[jj]]/df.loc[ii+12,'Discharge Capacity']
        
for jj in range(len(nfC_names)):
    idx = len(df)
    for ii in range(12):
        df.loc[idx+ii,:] = df.loc[ii+24,:]
        df.loc[idx+ii,'Elap_time'] = 21*(jj+1)
        df.loc[idx+ii,'Cap'] = dfn5.loc[ii+24,nfC_names[jj]]/df.loc[ii+24,'Discharge Capacity']
        
for jj in range(len(nfor_names)):
    idx = len(df)
    for ii in range(8):
        df.loc[idx+ii,:] = df.loc[ii+36,:]
        df.loc[idx+ii,'Elap_time'] = 21*(jj+1)
        df.loc[idx+ii,'Cap'] = dfn40.loc[ii+36,nfor_names[jj]]/df.loc[ii+36,'Discharge Capacity']

df
```
