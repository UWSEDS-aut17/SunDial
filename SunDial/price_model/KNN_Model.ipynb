```python
%matplotlib inline 
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.grid_search import GridSearchCV
from sklearn import cross_validation
from sklearn import grid_search
```

```python
from process_data import process_data
from process_data import get_date_range
```

```python
price_frame = process_data()
df_price_frame = price_frame.set_index("time")
df_price_frame.head()
```

```python
x_train, y_train, x_test, y_test = get_date_range(df_price_frame, '2016-01-01', '2016-03-01', '2016-03-01', '2016-03-05')
```

```python
neighbor_range = range(1, 30)

tuned_parameters = [{
    'n_neighbors': neighbor_range,
    }]

knn_price = GridSearchCV(KNeighborsRegressor(algorithm='auto'), param_grid=tuned_parameters, verbose=0)
y_pred = knn_price.fit(x_train, y_train).predict(x_test)

print 'Optimum parameters epsilon and kernel for KNN: ', knn_price.best_params_

print "The test score R2 for KNN: ", knn_price.score(x_test, y_test)

print("SVR mean squared error: %.2f"
      % np.mean((y_test - knn_price.predict(x_test)) ** 2))
```

```python
fig = plt.figure(figsize=(15,7))
plt.scatter(x_test.index, y_test, c='k', label='Observed')
plt.plot(x_test.index, y_pred, c='r', label='Predicted')
plt.xlabel('data')
plt.ylabel('lmp_value')
plt.title('KNN Regression')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
plt.scatter(y_test, y_test, c='k')
plt.scatter(y_test, y_pred, c='r')
plt.xlabel('Observed Elec. Usage (kWh)')
plt.ylabel("Predicted Elec. Usage (kWh): $\hat{Y}_i$")
plt.title("Energy vs Predicted Energy: $Y_i$ vs $\hat{Y}_i$")
```
